```python
## Korean fonts setting 
font_path = "C:\windows\Fonts\malgun.ttf"
## Get font property
font_prop = fm.FontProperties(fname = font_path).get_name()
## In Matplotlib, rc(run command) command >> set the korean font
matplotlib.rc('font', family = font_prop)
```

```python
## Create figure and plot
figure, (axes1, axes2) = plt.subplots(nrows= 2, ncols= 1)
## Figure size 
figure.set_size_inches(6,3)
print(figure)
print(axes1,axes2)

# seoul_df = data.loc[data['광역시도'] == '서울특별시',:]
# sns.barplot(data = seoul_df, x = '행정구역', y = '인구수', ax= axes1)
# sns.barplot(data = seoul_df.sort_values(by = '인구수', ascending = True), x = '행정구역', y = '인구수', ax= axes2)

```

```python
## Make a function 
def show_bar_plot(city, y_axis):
    city_df = data.loc[data['광역시도'] == city,:]
    # sns.barplot(data = city_df, x = '행정구역', y = value, ax= axes1)
    figure, axes = plt.subplots(nrows=1, ncols= 1)
    figure.set_size_inches(9,6)
    sns.barplot(data = city_df.sort_values(by = y_axis, ascending = True), x = '행정구역', y = y_axis, ax= axes)
show_bar_plot('부산광역시', '면적')
show_bar_plot('서울특별시', '인구수')
```
