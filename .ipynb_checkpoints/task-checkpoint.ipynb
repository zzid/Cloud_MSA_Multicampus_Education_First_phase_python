```python
import pandas as pd
data = pd.read_csv('docs/data/data_draw_korea.csv')
print(data.head())

print(data.tail())

print(data['인구수'].max())
print(data['인구수'].min())
print(data['인구수'].mean())

print(data['광역시도'].unique())

print(data.groupby('광역시도').count())

print(data.loc[data['광역시도'] == '서울특별시'].sort_values(by ='인구수', ascending = False ))
```

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin

url = 'https://news.naver.com/'
res = requests.get(url)

print(f'status code : {res.status_code}')

html_doc = res.text

soup = BeautifulSoup(html_doc, 'html.parser')

tag_list = soup.select("a[href*='read.nhn']")

# print(tag_list)
for i in tag_list:
    ttl = i.get('title')
    if ttl is None:
        continue
    link = urljoin(url, i['href'])
    print(f'{ttl}\n{link}')
```
